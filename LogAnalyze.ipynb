```python
import pandas as pd
import datetime as dt
from datetime import datetime

def getCSV(file):
    return pd.read_csv(file, engine='python')

def getDuplicate(column_name, df_data): 
    return df_data.drop_duplicates(subset = column_name)

def saveCSV(df, fname):
    pdf = pd.DataFrame(df) #データフレームをpandasに変換
    pdf.to_csv(fname)
```

```python
df_data = getCSV('./csv/duration_all.csv')
```

```python
df_ap = getCSV('./csv/APlocations.csv')
```

```python
aps = df_ap.AP.values.tolist()
aps.append('out')
print(aps[0:10])
print(len(aps))
```

```python
print(df_data.head())
```

```python
df_unique = getDuplicate('clientid', df_data)
```

```python
print(df_unique.head())
client = df_unique['clientid'].values.tolist() #クライアントリスト
print(client[0:10])
print(len(client))
```

```python
visit = [[0 for i in aps] for i in client] #APを訪れた回数
staytime = [[0 for i in aps] for i in client] #APに滞在した時間
stayhour = [[0 for i in range(24)] for i in client] #時間別の訪問AP回数 
stayday = [[0 for i in range(32)] for i in client] #日別の訪問AP回数
```

```python
idx = -1
clientid = ''
for index, row in df_data.iterrows():
    tmp_id = row['clientid']
    if(clientid != tmp_id): #clientが変更された時
        idx += 1
        print( str(idx + 1) + '/' + str(len(client)))
    from_time = dt.datetime.strptime(row['from_time'], '%Y-%m-%d %H:%M:%S')
    #print(from_time)
    #print(from_time.hour)
    #print(from_time.day)
    if(row['duration'] >= 0 and row['duration'] <= 3600 * 3):
        #print(row['duration'])
        visit[idx][row['from']] += 1
        staytime[idx][row['from']] += row['duration']#利用時間
        stayhour[idx][from_time.hour] += 1
        stayday[idx][from_time.day] += 1    
    
    #print(row['clientid'])
    #print(row['from_time'])
    #print(row['duration'])
    #print(row['from'])
    #print(row['to'])
    clientid = tmp_id
```

```python
saveCSV(visit, "./csv/visit.csv")
saveCSV(staytime, "./csv/staytime.csv")
saveCSV(stayhour, "./csv/stayhour.csv")
saveCSV(stayday, "./csv/stayday.csv")
```
